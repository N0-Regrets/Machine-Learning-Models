```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from LinearRegression import LinearRegression as MyLinearRegression
from sklearn.linear_model import LinearRegression

# to make this notebook's output stable across runs
np.random.seed(42)
```

```python
# Generate linear data
m = 100  # number of instances
n = 1  # number of features
X = 2 * np.random.rand( m , n) 
y = 4 + 3 * X + np.random.randn( m , n)  # linear function plus Gaussian noise

# Plot the data
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Create and fit my model
my_model = MyLinearRegression()
my_model.fit(X, y)
y_predict = my_model.predict(X)
```

```python
# Plot the data with the fitted line
plt.scatter(X, y )
plt.plot(X, y_predict, color="red")
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Create and fit the model form scikit-learn
sklearn_model = LinearRegression()
sklearn_model.fit(X, y)
```

```python
# print the weights of both models
print(sklearn_model.intercept_ , sklearn_model.coef_,"\n")
print(my_model.theta)
```
